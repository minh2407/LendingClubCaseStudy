```python
loan_df = pd.read_csv('loan.csv')
```

```python
# Can see that there's no headers or summary row presents from above dataframe. check the footer rows instead.
loan_df.tail()
```

```python
# Check if there's any unnecessary rows or not 
loan_df['loan_status'].value_counts()
```

```python
total_rows = len(loan_df)
total_rows
```

```python
# Drop the rows for loan_status="current" since the currently loan in progress cannot contribute to the conclusion
loan_df = loan_df[loan_df['loan_status']!= 'Current']
print('total number of rows at the moment', len(loan_df))
```

```python
# dang co van de
rows_deleted = ((total_rows - len(loan_df))*100)/total_rows
print('% of rows deleted', round(rows_deleted,2))
```

```python
# Check duplicate rows 
duplicate_rows = len(loan_df[loan_df.duplicated()])
print ('count of duplicate rows:', round(duplicate_rows,2))
```
